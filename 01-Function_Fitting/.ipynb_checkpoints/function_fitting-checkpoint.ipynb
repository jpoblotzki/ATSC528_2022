```python
### Import Required Modules ###
import numpy as np                 #numpy for math
import pandas as pd                #pandas because they're cute
import matplotlib.pyplot as plt    #matplotlib for plotting
import cartopy.crs as ccrs         #cartopy for plotting on map
import cartopy.feature as cfeature #cartopy basic shapefiles
```

```python
### define variables ###
xo = 18.9
yo = -6.3
delta_x = 1.27
delta_y = 1.27
phi_o = 60
lambda_o = -115
rho = 637100000. ## convert from km to cm
map_scale = 1/15000000

### Read in observations ###
file = open("C:\\Users\\julia\\Documents\\GitHub\\ATSC528_2022\\01-Function_Fitting\\RAOBs_201903131200.txt", "r")
data= pd.read_csv(file, sep=",")
data[data.columns[1:5]]=data[data.columns[1:5]].astype(float) ## make text to numbers
```

```python
### Set up analysis map with a 22x28 rectangular grid of points ###

## make a grid
l = xo+np.arange(22)*delta_x
k = yo+np.arange(28)*delta_y
x_grid, y_grid = np.meshgrid(l,k)

## convert grid from x,y to lat,lon for plotting purposes
grid_lon = np.arctan(y_grid/x_grid)*(180/np.pi)+lambda_o
grid_lat=(180/np.pi)*((np.pi/2)-(2*np.arctan(np.sqrt((x_grid/map_scale)**2+(y_grid/map_scale)**2)/(rho*(1+np.sin(phi_o*np.pi/180.))))))
```

```python
### convert obs lat/long to x,y (may want to plot on your analysis grid to verify)###

## convert obs from lat,lon to x,y
sigma = ((1.0+(np.sin(phi_o * np.pi / 180.))) / (1.0+(np.sin(data['Lat'] * np.pi / 180.))))
x = sigma * rho * np.cos(data['Lat'] * np.pi / 180.) * np.cos((data['Lon']-lambda_o) * np.pi / 180.) * map_scale
y = sigma * rho * np.cos(data['Lat'] * np.pi / 180.) * np.sin((data['Lon']-lambda_o) * np.pi / 180.) * map_scale

# make our obs x & y into an array for easier math-ing
x = np.array(x)
y = np.array(y)

## convert obs x,y back to lat,lon to plot
data_lon = np.arctan(y/x)*(180/np.pi)+lambda_o
data_lat=(180/np.pi)*((np.pi/2)-(2*np.arctan(np.sqrt((x/map_scale)**2+(y/map_scale)**2)/(rho*(1+np.sin(phi_o*np.pi/180.))))))

proj = ccrs.Stereographic(central_longitude=-115,central_latitude=90,true_scale_latitude=60)
fig = plt.figure(figsize=(8,8),dpi=200)
ax1 = fig.add_subplot(111,projection=proj)
ax1.add_feature(cfeature.STATES)
ax1.add_feature(cfeature.COASTLINE)
ax1.scatter(grid_lon.ravel(), grid_lat.ravel(),transform = ccrs.PlateCarree())
ax1.scatter(data_lon, data_lat, transform = ccrs.PlateCarree())

#plt.savefig("C:\\Users\\julia\\Documents\\GitHub\\ATSC528_2022\\01-Function_Fitting\\Obs_and_Grid_Locations.png")

plt.show()
```

```python
### Perform 500mb geopotential height analyses using a second order 2-d polynomial with two ###
### radii of influence (10cm & 20cm) ###

## define radius of influence matrix
ROI_arr = np.array([10,20,6])

## empty matrices to be filled later
y_n=22
x_n=28
analysis_mat = np.empty((x_n, y_n, len(ROI_arr))) ##for the values we calculate
obs_matrix = np.empty((x_n, y_n, len(ROI_arr))) ##for the number of obs being used

# create a nasty for loop to fill these matrices
for i in range(len(ROI_arr)):
    roi = ROI_arr[i]
    for m in range(len(x_grid)):
        for n in range(len(x_grid[0])):
            rel_x = x_grid[m,n] - x ## subtract your obs from your analysis to get...
            rel_y = y_grid[m,n] - y ## ...relative distance for our map
            radius = (rel_x**2 + rel_y**2)**(1/2) ## radius of influence for ith analysis point
            index_roi = np.where(radius <= roi)[0] ## only these obs will affect analysis value
            
            ## make obs matrix same shape and size as index_roi matrix
            obs_matrix[m,n,i] = len(index_roi)
                        
            ## create matrices for your f_o(x_k,y_k) using only values within radius of influence
            x_k = rel_x[index_roi] #x_k values within our R_i for each point
            y_k = rel_y[index_roi] #y_k values only within index_roi for each grid point
            fo = data['Height'][index_roi] #500-mb height values only within index_roi values
            fo = np.array(fo) ##do this due to those black & white fluffballs
            
            ## more empty matrices for C=R^-1*O
            R = np.zeros((6,6))
            O = np.zeros(6)
            
            ## fill in R matrix
            for q in range(len(y_k)):
                r2 = np.matrix([1,x_k[q],y_k[q],x_k[q]**2,y_k[q]**2,x_k[q]*y_k[q]])
                r1 = r2.T
                r = r1 * r2
                R += np.array(r)
                O += np.array([fo[q],x_k[q]*fo[q],y_k[q]*fo[q],x_k[q]**2*fo[q],y_k[q]**2*fo[q],x_k[q]*y_k[q]*fo[q]])
                
            ## get the average values by dividing by total number of obs used
            R = R/obs_matrix[m,n,i]
            O = O/obs_matrix[m,n,i]
            
            ## calculate C now
            C = np.linalg.inv(np.matrix(R))*np.matrix(O).T
            analysis_mat[m,n,i] = C[0,0] #set equal to C_00 due to that being the only value we care about
```

```python
### Plot 500mb analyses over a map ###
#convert analysis x,y to lat/lon#

proj = ccrs.Stereographic(central_longitude=-115,central_latitude=90,true_scale_latitude=60)
fig = plt.figure(figsize=(8,8),dpi=200)
ax1 = fig.add_subplot(111,projection=proj)
ax1.add_feature(cfeature.STATES)
ax1.add_feature(cfeature.COASTLINE)
## plot grid lat & lons because we want analysis locations NOT obs locations
cs1 = ax1.contour(grid_lon,grid_lat,analysis_mat[:,:,0],colors='k',levels=np.arange(0,8000,60),transform=ccrs.PlateCarree())
plt.clabel(cs1,levels=np.arange(0,8000,60))

#plt.savefig("C:\\Users\\julia\\Documen ts\\GitHub\\ATSC528_2022\\01-Function_Fitting\\Heights_at_ROI_10cm.png")
#plt.savefig("C:\\Users\\julia\\Documents\\GitHub\\ATSC528_2022\\01-Function_Fitting\\Heights_at_ROI_20cm.png")
#plt.savefig("C:\\Users\\julia\\Documents\\GitHub\\ATSC528_2022\\01-Function_Fitting\\Heights_at_ROI_6cm.png")

plt.show()
```

```python
### Plot number of observations available to polynomial fitting scheme for each grid point ###
proj = ccrs.Stereographic(central_longitude=-115,central_latitude=90,true_scale_latitude=60)
fig = plt.figure(figsize=(8,8),dpi=200)
ax1 = fig.add_subplot(111,projection=proj)
ax1.add_feature(cfeature.STATES)
ax1.add_feature(cfeature.COASTLINE)
#plot number of observations used at each grid point
cs1 = ax1.contour(grid_lon,grid_lat,obs_matrix[:,:,0],colors='k',levels=np.arange(0,200,5),transform=ccrs.PlateCarree())
plt.clabel(cs1,levels=np.arange(0,200,10))

#plt.savefig("C:\\Users\\julia\\Documents\\GitHub\\ATSC528_2022\\01-Function_Fitting\\Obs_Available_at_ROI_10cm.png")
#plt.savefig("C:\\Users\\julia\\Documents\\GitHub\\ATSC528_2022\\01-Function_Fitting\\Obs_Available_at_ROI_20cm.png")
#plt.savefig("C:\\Users\\julia\\Documents\\GitHub\\ATSC528_2022\\01-Function_Fitting\\Obs_Available_at_ROI_6cm.png")

plt.show()
```
